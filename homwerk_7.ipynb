```python
from sklearn.cluster import KMeans

km = KMeans (n_clusters=4)

km
```

```python
y_pred= km.fit_predict(df[[ 'a1', 'a2','a3','a4','a5','a6','a7']])
y_pred
```

```python
df['Cluster']=y_pred 
df
```

```python
km.cluster_centers_
```

```python
df1=df[df.Cluster==0]
df2=df[df.Cluster==1]
df3=df[df.Cluster==2]
df4=df[df.Cluster==3]
plt.scatter(df1.a1, df1['a1'], color='green', marker='*', label='a1')
plt.scatter(df2.a2, df2[ 'a2'], color='red', marker='*', label='a2')
plt.scatter(df3.a3, df3[ 'a3'], color='black', marker='*', label='a3')
plt.scatter(df4.a4, df4['a4'], color='blue', marker='*', label='a4')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple', marker='*', label='centroid')
plt.xlabel('a1')
plt.ylabel('a2')
plt.legend()
```
